```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame from the extracted data
data = {
    "Epoch": list(range(1, 11)),
    "Training Loss": [1.5634, 1.2276, 1.0130, 1.0363, 0.7323, 0.6083, 0.4541, 0.3570, 0.2355, 0.1585],
    "Validation Loss": [1.2959, 6.0193, 1.1357, 1.0363, 1.0239, 1.0151, 1.2793, 1.2550, 1.4289, 1.7127],
    "Accuracy": [0.4568, 0.4259, 0.6111, 0.6728, 0.5988, 0.7037, 0.7222, 0.7346, 0.7346, 0.7469]
}

df = pd.DataFrame(data)

# Plot Training Loss Curve
plt.figure(figsize=(8, 5))
plt.plot(df["Epoch"], df["Training Loss"], marker="o", linestyle="-", color="blue", label="Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.xticks(df["Epoch"])  # Ensure x-axis increments by 1
plt.title("Training Loss Curve")
plt.legend()
plt.grid(True)
plt.show()

# Plot Validation Loss Curve
plt.figure(figsize=(8, 5))
plt.plot(df["Epoch"], df["Validation Loss"], marker="s", linestyle="-", color="red", label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.xticks(df["Epoch"])  # Ensure x-axis increments by 1
plt.title("Validation Loss Curve")
plt.legend()
plt.grid(True)
plt.show()

# Plot Accuracy Curve
plt.figure(figsize=(8, 5))
plt.plot(df["Epoch"], df["Accuracy"], marker="o", linestyle="-", color="green", label="Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.xticks(df["Epoch"])  # Ensure x-axis increments by 1
plt.title("Accuracy Curve")
plt.legend()
plt.grid(True)
plt.show()
```
