```python
import pandas as pd
import matplotlib.pyplot as plt

# Data matching the provided screenshot
training_data = {
    "Epoch": list(range(1, 11)),
    "Training Loss": [1.3261, 1.0467, 0.9252, 0.7061, 0.5618, 0.4417, 0.3211, 0.2547, 0.1967, 0.2632],
    "Validation Loss": [1.0533, 1.1768, 1.4028, 0.9877, 1.0498, 1.2203, 1.0878, 1.1357, 1.0942, 1.0463],
    "Accuracy": [0.5986, 0.7291, 0.7223, 0.7865, 0.8001, 0.8372, 0.8469, 0.8537, 0.8479, 0.8421]
}

df = pd.DataFrame(training_data)

# Plot Training Loss
plt.figure(figsize=(8, 5))
plt.plot(df["Epoch"], df["Training Loss"], marker="o", linestyle="-", color="blue", label="Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.xticks(df["Epoch"])
plt.title("Training Loss Curve")
plt.legend()
plt.grid(True)
plt.show()

# Plot Validation Loss
plt.figure(figsize=(8, 5))
plt.plot(df["Epoch"], df["Validation Loss"], marker="s", linestyle="-", color="red", label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.xticks(df["Epoch"])
plt.title("Validation Loss Curve")
plt.legend()
plt.grid(True)
plt.show()

# Plot Accuracy
plt.figure(figsize=(8, 5))
plt.plot(df["Epoch"], df["Accuracy"], marker="o", linestyle="-", color="green", label="Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.xticks(df["Epoch"])
plt.title("Accuracy Curve")
plt.legend()
plt.grid(True)
plt.show()
```
